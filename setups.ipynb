### Setup Definition files

```python
%sql
USE CATALOG bridgestore;
```

```python
%sql

-----CRIAR CATALOGO COM NOSSO EXTERNAL LOCATION 
CREATE CATALOG bridgestore
MANAGED LOCATION 'abfss://brigdestorage@airbytestorage35.dfs.core.windows.net/'
COMMENT 'Descrição opcional do meu catálogo';
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS bridgestore.bronze
COMMENT "Camada BRONZE - dados padronizados, com metadados e controle de ingestão";

CREATE SCHEMA IF NOT EXISTS brigdestore.silver
COMMENT "Camada SILVER - dados tratados, com regras de negócio aplicadas";

CREATE SCHEMA IF NOT EXISTS bridgestore.gold
COMMENT "Camada GOLD - dados analíticos e métricas finais para BI e IA";
```

```python
%sql
SHOW SCHEMAS IN bridgestore;
```

```python
%sql
-- RAW (restrito a engenheiros)
GRANT USE SCHEMA ON SCHEMA bridgestore.airbyte_internal TO data_engineers;
-- GRANT SELECT ON ALL TABLES IN SCHEMA lakehouse.raw TO data_engineers;
```

```python
%sql
-- Catálogo
GRANT USE CATALOG ON CATALOG bridgestore TO data_engineers;
GRANT USE CATALOG ON CATALOG bridgestore TO data_analysts;
GRANT USE CATALOG ON CATALOG bridgestore TO data_scientists;
```

```python
%sql
---------------------------
-- 1) BRONZE (dados brutos)
---------------------------

-- Engenheiros: acesso total (criar, ler, alterar, deletar)
GRANT USAGE ON SCHEMA bridgestore.bronze TO data_engineers;
GRANT CREATE, SELECT, INSERT, UPDATE, DELETE ON SCHEMA bridgestore.bronze TO data_engineers;
GRANT SELECT, INSERT, UPDATE, DELETE ON ALL TABLES IN SCHEMA bridgestore.bronze TO data_engineers;

-- Analytics e Scientists: somente leitura
GRANT USAGE, SELECT ON SCHEMA bridgestore.bronze TO data_analytics;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.bronze TO data_analytics;

GRANT USAGE, SELECT ON SCHEMA bridgestore.bronze TO data_scientists;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.bronze TO data_scientists;



---------------------------
-- 2) SILVER (dados limpos)
---------------------------

-- Engenheiros: acesso total
GRANT USAGE ON SCHEMA bridgestore.silver TO data_engineers;
GRANT CREATE, SELECT, INSERT, UPDATE, DELETE ON SCHEMA bridgestore.silver TO data_engineers;
GRANT SELECT, INSERT, UPDATE, DELETE ON ALL TABLES IN SCHEMA bridgestore.silver TO data_engineers;

-- Analytics e Scientists: leitura
GRANT USAGE, SELECT ON SCHEMA bridgestore.silver TO data_analytics;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.silver TO data_analytics;

GRANT USAGE, SELECT ON SCHEMA bridgestore.silver TO data_scientists;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.silver TO data_scientists;



---------------------------
-- 3) GOLD (camada final)
---------------------------

-- Engenheiros: leitura (não devem sobrescrever dashboards)
GRANT USAGE, SELECT ON SCHEMA bridgestore.gold TO data_engineers;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.gold TO data_engineers;

-- Analytics: leitura — principal camada de consumo
GRANT USAGE, SELECT ON SCHEMA bridgestore.gold TO data_analytics;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.gold TO data_analytics;

-- Scientists: leitura (dados já refinados)
GRANT USAGE, SELECT ON SCHEMA bridgestore.gold TO data_scientists;
GRANT SELECT ON ALL TABLES IN SCHEMA bridgestore.gold TO data_scientists;
```
